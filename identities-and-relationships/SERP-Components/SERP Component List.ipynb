```python
from collections import OrderedDict
with open("Copy of SERP Analysis - Merged data.csv") as fileToRead:     # open file to read
    csvReader = csv.DictReader(fileToRead)   # create reader object
    rows = [row for row in csvReader]        # do the reading of each row

componentResults = []

categories = rows[0].keys()
#print(categories)
for category in categories:
    if category == 'queries':
        continue
    #print(category)
    for row in rows:    
        #print(row[category])
        if row[category] == 'Y':
            entry = OrderedDict({'Component': category, 
                     'Query': row['queries']})
            #print(entry)
            componentResults.append(entry)
            
print(componentResults)
```

```python
with open("componentResults.csv", 'w') as fileToWrite:
    csvWriter = csv.DictWriter(fileToWrite, 
                               fieldnames=componentResults[0].keys()) # get the key names as column names
    
    csvWriter.writeheader() # write the column names
    csvWriter.writerows(componentResults) # write all rows
```
